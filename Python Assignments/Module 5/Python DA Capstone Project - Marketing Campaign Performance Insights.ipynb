# 3) Exploratory Data Analysis (EDA) and Visualization:

## Campaign Performance:


```python
#Plot a scatter plot to visualize the relationship between Acquisition_Cost and ROI.

import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=df, x='Acquisition_Cost', y='ROI', hue='Campaign_Type')
plt.title('Relationship between Acquisition Cost and ROI')
plt.show()
```

```python
# Create a bar chart to visualize the average Conversion_Rate for different Channel_Used, categorized by Campaign_Type.

import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size before plotting
plt.figure(figsize=(7,4))

sns.barplot(data=df, x='Channel_Used', y='Conversion_Rate', hue='Campaign_Type')

plt.title('Average Conversion Rate by Channel and Campaign Type')
plt.xlabel('Channel Used')
plt.ylabel('Average Conversion Rate')
plt.legend(title='Campaign Type', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# Visualize the distribution of Engagement_Score across different Campaign_Type using a box plot.

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))  # set figure size

sns.boxplot(data=df, x='Campaign_Type', y='Engagement_Score', palette='pastel', hue='Channel_Used')

plt.title('Distribution of Engagement Score across Campaign Types')
plt.xlabel('Campaign Type')
plt.ylabel('Engagement Score')
plt.legend(title='Channel_Used', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Analyze the average ROI by Company using a bar chart to compare the profitability of campaigns conducted by different companies.

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))

# Bar chart showing average ROI per company
sns.barplot(
    data=df,
    x='Company',
    y='ROI',
    estimator='mean',
    palette='coolwarm', hue = 'Company'
)

plt.title('Average ROI by Company')
plt.xlabel('Company')
plt.ylabel('Average ROI')

plt.xticks(rotation=45)  # rotate names if long
plt.tight_layout()
plt.show()
```

```python
# Examine the correlation between Engagement_Score and Conversion_Rate using a heatmap.

import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation
corr = df[['Engagement_Score', 'Conversion_Rate']].corr()

# Plot heatmap
plt.figure(figsize=(5, 4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')

plt.title('Correlation between Engagement Score and Conversion Rate')
plt.show()
```

Customer Segmentation:

```python
# Create a count plot to visualize the distribution of Target_Audience.

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(7, 5))

sns.countplot(data=df, x='Target_Audience', hue='Campaign_Type', palette='pastel')

plt.title('Distribution of Target Audience')
plt.xlabel('Target Audience')
plt.ylabel('Count')

plt.xticks(rotation=45)  # rotate if names are long
plt.tight_layout()
plt.show()
```

```python
# Identify which Customer_Segment has the highest Conversion_Rate for each Language using a bar chart.

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))

sns.barplot(
    data=df,
    x='Language',
    y='Conversion_Rate',
    hue='Customer_Segment',  # 👈 comparison by segment
    estimator='mean',
    palette='Set2'
)

plt.title('Average Conversion Rate by Language and Customer Segment')
plt.xlabel('Language')
plt.ylabel('Average Conversion Rate')

plt.legend(title='Customer Segment', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Visualize the distribution of Acquisition_Cost across each Customer_Segment, categorized by Channel_Used, using a box plot.

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

sns.boxplot(data=df,x='Customer_Segment',y='Acquisition_Cost',hue='Channel_Used',palette='Set2')

plt.title('Distribution of Acquisition Cost by Customer Segment and Channel Used')
plt.xlabel('Customer Segment')
plt.ylabel('Acquisition Cost')

plt.legend(title='Channel Used', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Analyze average Conversion_Rate by Language using a bar chart to compare the effectiveness of campaigns conducted in different languages.

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(9, 5))

# Create bar chart for average Conversion_Rate by Language
sns.barplot(
    data=df,
    x='Language',
    y='Conversion_Rate',
    estimator='mean',      # average value
    palette='Set2', hue = 'Channel_Used'
)

plt.title('Average Conversion Rate by Language')
plt.xlabel('Language')
plt.ylabel('Average Conversion Rate')

plt.xticks(rotation=50)  # rotate if language names are long
plt.tight_layout()
plt.show()
```

### Channel Effectiveness:

```python
# Compare the Engagement_Score for different Channels_Used, segmented by Campaign_Type, using a bar chart.

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))

sns.barplot(
    data=df,
    x='Channel_Used',
    y='Engagement_Score',
    hue='Campaign_Type',       # 👈 segmentation by campaign type
    estimator='mean',
    palette='Set2'
)

plt.title('Average Engagement Score by Channel Used and Campaign Type')
plt.xlabel('Channel Used')
plt.ylabel('Average Engagement Score')

plt.legend(title='Campaign Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Show the distribution of total ROI across different Channels_Used using a pie chart.

import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Group data by Channel_Used and sum the ROI
roi_by_channel = df.groupby('Channel_Used')['ROI'].sum()

# Step 2: Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(
    roi_by_channel,
    labels=roi_by_channel.index,     # channel names
    autopct='%1.1f%%',               # show percentages
    startangle=90,                   # start from top
    colors=plt.cm.Set2.colors        # nice color palette
)

plt.title('Distribution of Total ROI across Channels Used')
plt.show()
```

```python
# Plot a scatter plot to show the relationship between Clicks and Impressions for each Channel_Used.

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(7, 5))

sns.scatterplot(
    data=df,
    x='Impressions',
    y='Clicks',
    hue='Channel_Used',      # 🎨 color by channel
    style='Channel_Used',    # 🌀 optional: different marker shapes
    palette='Set2',
    s=80,                    # point size
    alpha=0.7                # transparency for overlap
)

plt.title('Relationship between Clicks and Impressions by Channel Used')
plt.xlabel('Impressions')
plt.ylabel('Clicks')
plt.legend(title='Channel Used', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Time -Based Analysis:

```python
# Plot the distribution of Duration using a histogram.

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(7, 5))

sns.histplot(
    data=df,
    x='Duration',
    bins=20,          # number of bins (you can adjust)
    kde=True,         # adds a smooth curve
    color='skyblue'
)

plt.title('Distribution of Campaign Duration')
plt.xlabel('Duration (days)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Analyze how the overall Conversion_Rate has changed over Date for each Company using a line chart.

import seaborn as sns
import matplotlib.pyplot as plt

# Ensure 'Date' column is in datetime format
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

plt.figure(figsize=(10, 6))

sns.lineplot(
    data=df,
    x='Date',
    y='Conversion_Rate',
    hue='Company',         # 🎨 one line per company
    marker='o',            # 🟢 show points
    palette='Set2'
)

plt.title('Trend of Conversion Rate over Time for Each Company')
plt.xlabel('Date')
plt.ylabel('Conversion Rate')
plt.legend(title='Company', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Examine the trend of Engagement_Score over Date with a line chart.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Ensure 'Date' column is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Sort the dataset by date (important for a clean line)
df = df.sort_values('Date')

plt.figure(figsize=(10, 6))

sns.lineplot(
    data=df,
    x='Date',
    y='Engagement_Score',
    marker='o',          # show data points
    color='royalblue'    # set color (optional)
)

plt.title('Trend of Engagement Score over Time')
plt.xlabel('Date')
plt.ylabel('Engagement Score')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### Geographic Analysis:

```python
# Determine which location has the highest Acquisition_Cost using a bar chart.

import seaborn as sns
import matplotlib.pyplot as plt

# Group data by Location and calculate the average Acquisition_Cost
avg_cost_by_location = df.groupby('Location')['Acquisition_Cost'].mean().reset_index()

plt.figure(figsize=(8, 6))

sns.barplot(
    data=avg_cost_by_location,
    x='Location',
    y='Acquisition_Cost',
    palette='Set2', hue = 'Location'
)

plt.title('Average Acquisition Cost by Location')
plt.xlabel('Location')
plt.ylabel('Average Acquisition Cost')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Visualize the Conversion_Rate by different Location, categorized by Target_Audience, using a bar chart.

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

sns.barplot(
    data=df,
    x='Location',
    y='Conversion_Rate',
    hue='Target_Audience',     # 🎨 separate bars by audience
    estimator='mean',          # use average conversion rate
    palette='Set2'
)

plt.title('Conversion Rate by Location and Target Audience')
plt.xlabel('Location')
plt.ylabel('Average Conversion Rate')
plt.xticks(rotation=45)
plt.legend(title='Target Audience', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Illustrate the proportion of ROI by Location using a pie chart.

import matplotlib.pyplot as plt

# Step 1: Group ROI by Location
roi_by_location = df.groupby('Location')['ROI'].sum()

# Step 2: Create pie chart
plt.figure(figsize=(7, 7))

explode = [0.1 if val == roi_by_location.max() else 0 for val in roi_by_location]
plt.pie(roi_by_location, labels=roi_by_location.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Set2.colors, explode=explode)



plt.title('Proportion of ROI by Location')
plt.tight_layout()
plt.show()
```
