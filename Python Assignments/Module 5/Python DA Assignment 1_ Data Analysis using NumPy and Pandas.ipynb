Numpy Array Operations:
1. Create a 1D NumPy Array:

```python
import numpy as np

# Week 1 temperatures
temperatures_w1 = np.array([22.5, 25.3, 20.8, 23.4, 26.1, 24.8, 21.9])

# Display the array
print("Week 1 Temperatures:", temperatures_w1)
```

2. Inspection and Properties:

```python
# Inspect shape
print("Shape of the array:", temperatures_w1.shape)

# Inspect data type
print("Data type of the array:", temperatures_w1.dtype)

# Inspect number of elements
print("Number of elements:", temperatures_w1.size)
```

3. Array Operations:

```python
# Convert to Fahrenheit
temperatures_w1_fh = (temperatures_w1 * 9/5) + 32

# Display the converted array
print("Week 1 Temperatures in Fahrenheit:", temperatures_w1_fh)
```

```python
# Maximum temperature
max_temp = np.max(temperatures_w1)

# Minimum temperature
min_temp = np.min(temperatures_w1)

# Mean (average) temperature
mean_temp = np.mean(temperatures_w1)

# Display results
print("Maximum Temperature:", max_temp, "°C")
print("Minimum Temperature:", min_temp, "°C")
print("Mean Temperature:", round(mean_temp, 2), "°C")
```

4. Array Slicing and Indexing:

```python
# Extract temperatures for first 3 days (index 0, 1, 2)
first_three_days = temperatures_w1[:3]

# Display result
print("Temperatures for the first three days:", first_three_days)
```

```python
# Extract temperatures for Weekend (2 days)
Weekend_days = temperatures_w1[-2:]

# Display result
print("Temperatures for the Weekends:", Weekend_days)
```

```python
# Extract temperatures for middle  (3 days)
middle_days = temperatures_w1[2:5]

# Display result
print("Temperatures for the Middle days:", middle_days)
```

5. Create a 2D Array:

```python
# Week 2 temperatures
temperatures_w2 = np.array([19.2, 22.5, 21.3, 24.0, 23.5, 22.8, 20.1])

# Display the array
print("Week 1 Temperatures:", temperatures_w1)
print("Week 2 Temperatures:", temperatures_w2)
```

6. Inspect and Slice the 2D Array:

```python
# Combine into a 2D array
temperatures = np.array([temperatures_w1, temperatures_w2])

# Inspect shape
print("Shape of the array:", temperatures.shape)

# Inspect data type
print("Data type of the array:", temperatures.dtype)

# Inspect number of elements
print("Number of elements:", temperatures.size)
```

```python
# Extract Week 1 (row 0) & temperatures is a 2D array
week1 = temperatures[0]
print("Week 1 Temperatures:", week1)

# Extract Week 2 (row 1)
week2 = temperatures[1]
print("Week 2 Temperatures:", week2)

# Extract Weekend (last two days → columns 5 and 6)
weekends = temperatures[:, -2:]  # All rows, last 2 columns
print("Weekend Temperatures (Both Weeks):\n", weekends)
```

Pandas Series:
1. Creating Pandas Series:

```python
import pandas as pd

# Create the Series
marks = pd.Series(
    [95, 92, 89, 85, 80],
    index=['Rank1', 'Rank2', 'Rank3', 'Rank4', 'Rank5'])

# Display the Series
print(marks)
```

2. Indexing and Slicing:

```python
print("Mark of 1st Rank Student:", marks[0])
```

```python
# Retrieve top 3 ranks using index labels
top3_marks = marks.loc[['Rank1', 'Rank2', 'Rank3']]

print("Marks of Top 3 Ranks:\n", top3_marks)
```

```python
# Retrieve mark of the 3rd rank student (index position 2)
third_rank_mark = marks.iloc[2]

print("Mark of 3rd Rank Student:", third_rank_mark)
```

```python
# Apply boolean mask to filter marks > 90
high_scorers = marks[marks > 90]

print("Ranks with marks greater than 90:\n", high_scorers)
```

3. Manipulating Series:

```python
# Modify the mark of 1st rank student
marks['Rank1'] = 100

print("Updated Marks Series:\n", marks)
```

```python
# Remove the last rank (Rank5)
marks = marks.drop('Rank5')

print("Marks Series after removing last rank:\n", marks)
```

```python
# Compute CGPA
cgpa = marks / 10

print("CGPA of Students:\n", cgpa)
```

Pandas DataFrame:
1. Creating Pandas DataFrame:

```python
import pandas as pd

# Data for the DataFrame
data = {
    'TransactionID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'ProductCategory': ['Electronics', 'Clothing', 'Electronics', 'Furniture',
                        'Clothing', 'Electronics', 'Furniture', 'Clothing',
                        'Furniture', 'Electronics'],
    'Region': ['North', 'South', 'North', 'East', 'West', 'North', 'East',
               'West', 'South', 'North'],
    'Amount': [200, 150, 300, 450, 200, 250, 300, 180, 350, 400]
}

# Create DataFrame
transactions = pd.DataFrame(data)

# Display the DataFrame
print(transactions)
```

2. Data Exploration:

```python
# Display the DataFrame
print("Transactions DataFrame:")
print(transactions)

# Display the first 5 rows
print("\nHead of the DataFrame:")
print(transactions.head())

# Display the last 5 rows
print("\nTail of the DataFrame:")
print(transactions.tail())

# Display the shape (rows, columns)
print("\nShape of the DataFrame:")
print(transactions.shape)

# Display column names
print("\nColumn Names:")
print(transactions.columns)

# Display data types of each column
print("\nData Types:")
print(transactions.dtypes)
```

```python
# Display only 'ProductCategory' and 'Amount' columns
selected_columns = transactions[['ProductCategory', 'Amount']]
print(selected_columns)
```

```python
# Retrieve last 3 columns using iloc
last_three_columns_iloc = transactions.iloc[:, -3:]
print(last_three_columns_iloc)
```

```python
# Filter rows where Region is 'North' and Amount > 200
filtered_rows = transactions[(transactions['Region'] == 'North') & (transactions['Amount'] > 200)]
print(filtered_rows)
```

```python
# Value counts for 'ProductCategory'
category_counts = transactions['ProductCategory'].value_counts()
print(category_counts)
```

```python
# Unique values in 'Region' column
unique_regions = transactions['Region'].unique()
print(unique_regions)
```

```python
# Group by 'Region' and calculate mean amount
mean_amount_by_region = transactions.groupby('Region')['Amount'].mean()
print(mean_amount_by_region)
```

```python
print(transactions)
```

3. Manipulating the DataFrame:

```python
# Update Amount for TransactionID 102
transactions.loc[transactions['TransactionID'] == 102, 'Amount'] = 165

# Display the updated DataFrame
print(transactions)
```

```python
# Add a new column 'Discount' as 10% of 'Amount'
transactions['Discount'] = transactions['Amount'] * 0.10

# Display the updated DataFrame
print(transactions)
```

```python
# Remove the row where TransactionID is 109
transactions = transactions[transactions['TransactionID'] != 109]

# Display the updated DataFrame
print(transactions)
```

```python
# Delete the 'Discount' column
transactions = transactions.drop(columns=['Discount'])

# Display the updated DataFrame
print(transactions)
```
