While loop & Control Statements :
Number Guessing Game


```python
import random

# Step 1: Set up the game
secret_number = random.randint(1, 10)
attempts = 3

print("Welcome to the Number Guessing Game!")
print("I'm thinking of a number between 1 and 10.")
print(f"You have {attempts} attempts to guess it!\n")

# Step 2: Guessing logic
while attempts > 0:
    guess = int(input("Enter your guess (1–10): "))

    # Out of range → skip rest of loop
    if guess < 1 or guess > 10:
        print("Your guess is out of range. Please guess a number between 1 and 10.\n")
        continue

    # Correct guess
    if guess == secret_number:
        print("Congratulations! You guessed the correct number!")
        break

    # Wrong guess
    elif guess > secret_number:
        print("Too high! Try a smaller number.\n")
    else:
        print("Too low! Try a bigger number.\n")

    # Decrease attempts
    attempts -= 1
    print(f"Attempts left: {attempts}\n")

# Step 3: If user didn’t guess correctly
else:
    print(f"Better luck next time! The number was {secret_number}.")
```

Multiplication Table Generator

```python
x = int(input("Enter the number for which you want the multiplication table : "))
for i in range(1,11):
  print (f"{x} x {i} = {x*i}")
```

Function:
BMI Calculator

```python
# Step 1: Define the function
def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)
    return bmi

# Step 2: Take user input
weight = float(input("Enter your weight in kg: "))
height = float(input("Enter your height in meters: "))

# Step 3: Call the function and display result
bmi_value = calculate_bmi(weight, height)
print(f"Your BMI is: {bmi_value:.2f}")
```
