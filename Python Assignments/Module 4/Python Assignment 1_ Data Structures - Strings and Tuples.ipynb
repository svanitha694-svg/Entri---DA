1. String Concatenation:

```python
name = input("Enter your Name:")
print('Hello,', name)
```

```python
greeting_message = "Hello," + name + "! Welcome to Python Programming"
print(greeting_message)
```

2. String Slicing and Indexing:

```python
text = greeting_message
print(text)
```

```python
text[0] #print first charater of the string
```

```python
text[-1] #print last charater of the string
```

```python
text[0:5] #print first 5 charater of the string
```

```python
text[-11:] #print last 11 charater of the string
```

```python
text[::-1] #print the string in reverse
```

```python
text[-18:-12]  # use slicing and print the word 'python' in the string
```
