List (Creation, Modification and Access):
1. List Creation:

```python
# create age_list and name_list
age_list = [24,25,26,27,28]
name_list = ['Abi', 'Bharathi', 'Divya', 'Elan', 'Jeni']
```

2. List Operations / Modifications:

```python
name_list.append('Yazhini') # append
name_list
```

```python
age_list.insert(2,30)  #insert the element by index
age_list
```

```python
name_list.remove('Yazhini')  # remove by 'Yazhini'
name_list
```

```python
age_list.pop() # remove last element
```

```python
age_list.extend([29,30,26])  # insert many elements at a time
age_list
```

```python
age_list.sort(reverse=True)  # sort age by descending
age_list
```

```python
max_age = max(age_list)  # maximum age
max_age
```

```python
min_age = min(age_list)  # minimum age
min_age
```

```python
sum_age = sum(age_list) # sum age
sum_age
```

Operators & Conditional Statements :
(IF, ELIF, ELSE)

```python
# Performance Category Program

# 1. Prompt user for input
score = int(input("Enter your score (0 to 10): "))

# 2. Find performance category
if score > 7 and score <= 10:
    print("Above Average: Excellent work! Keep pushing towards mastery.")
elif 4 <= score <= 7:
    print("Average: Good effort! Keep practicing, there's room for improvement.")
elif 0 <= score < 4:
    print("Below Average: Need to improve your performance, consistent practice will lead to better results.")
else:
    print("Invalid score! Please enter a number between 0 and 10.")
```
